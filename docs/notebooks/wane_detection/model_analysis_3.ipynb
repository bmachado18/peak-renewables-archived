```python
import open3d as o3d
import numpy as np
import copy
import matplotlib.pyplot as plt
```

```python
mesh_path = './models/Aric Board 3.stl'
mesh = o3d.io.read_triangle_mesh(mesh_path)
#mesh.compute_vertex_normals()
mesh.compute_triangle_normals()
mesh_coord_frame = o3d.geometry.TriangleMesh.create_coordinate_frame(size=50, origin=[0, 0, 0])

box = mesh.get_minimal_oriented_bounding_box(robust=True)
box.color = (1,0,0)
o3d.visualization.draw_geometries([box, mesh_coord_frame, mesh])
```

```python

box_mesh = mesh.create_box(box.extent[0], box.extent[1], box.extent[2])

mesh = mesh.translate((-1)*mesh.get_center())

# Step 2: rotating to align with x axis: assuming R is relative to center
R_inv = np.linalg.inv(box.R)
mesh.rotate(R_inv, center=(0, 0, 0))

# new test for Step 3
mesh = mesh.translate(box.extent*(0.5))



o3d.visualization.draw_geometries([box, box_mesh, mesh_coord_frame, mesh])
```

```python
# normals are vectors that are orthogonol to triangle faces. each point has an associated normal vector
# hypothesis: should be able to filter the mesh and only keep triangles that have faces in particular direction
print(help(mesh))
```

```python
test_mesh = o3d.t.geometry.TriangleMesh.from_legacy(mesh)

box_mesh.compute_vertex_normals()
box_mesh_t = o3d.t.geometry.TriangleMesh.from_legacy(box_mesh)

diff = test_mesh.boolean_difference(box_mesh_t)

o3d.visualization.draw([diff, mesh_coord_frame])
```
