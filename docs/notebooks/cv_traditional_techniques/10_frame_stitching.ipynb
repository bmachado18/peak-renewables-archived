```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import sys
```

```python
left_1 = cv2.imread('../images/stitching/1-3.png')
middle_1 = cv2.imread('../images/stitching/1-2.png')
right_1 = cv2.imread('../images/stitching/1-1.png')

plt.imshow(cv2.cvtColor(left_1,cv2.COLOR_BGR2RGB))
plt.title('left_1')
plt.show()

plt.imshow(cv2.cvtColor(middle_1,cv2.COLOR_BGR2RGB))
plt.title('middle_1')
plt.show()

plt.imshow(cv2.cvtColor(right_1,cv2.COLOR_BGR2RGB))
plt.title('right_1')
plt.show()
```

```python
imgs = [left_1, middle_1, right_1]
stitcher = cv2.Stitcher.create(cv2.stitcher_SCANS)
status, pano = stitcher.stitch(imgs)

if status != cv2.Stitcher_OK:
    print("Can't stitch images, error code = %d" % status)
    sys.exit(-1)

plt.imshow(cv2.cvtColor(right_1,cv2.COLOR_BGR2RGB))
plt.title('right_1')
plt.show()
```
